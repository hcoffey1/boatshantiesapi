```python
from SimulationEnvironment import SimulationEnvironment
from Globot import Globot
import matplotlib.pyplot as plt
import numpy as np
from random import choice
```

```python
env = SimulationEnvironment(world_size=100, degradation=1.05)
SIMULATION_NAME = "seeker_field_new_signal"
global_counter = 0
```

```python
def print_state(heatmap):
    global SIMULATION_NAMEf
    global global_counter
    plt.figure(figsize=(8,8))
    plt.axis('off')
    plt.imshow(heatmap, interpolation='nearest')
    plt.savefig("{}/{:04}.png".format(SIMULATION_NAME, global_counter))
    global_counter += 1
```

```python
print_state(env.get_heatmap())
print_state(env.get_heatmap())
```

```python
beacon = Globot.build_from_file("rulesets/beacon.json")
env.add_globot(beacon, 25, 25)
beacon = Globot.build_from_file("rulesets/beacon.json")
env.add_globot(beacon, 24, 25)
beacon = Globot.build_from_file("rulesets/beacon.json")
env.add_globot(beacon, 25, 24)
beacon = Globot.build_from_file("rulesets/beacon.json")
env.add_globot(beacon, 24, 24)
env.step()
print_state(env.get_heatmap())
```

```python
for x in range(5, 105, 20):
    for y in range(5, 105, 20):
        seeker = Globot.build_from_file("rulesets/seeker.json")
        seeker.noise_pr = 0.8
        seeker.noise_int = 3
        y_random = choice([-2, -1, 0, 1, 2])
        x_random = choice([-2, -1, 0, 1, 2])
        env.add_globot(seeker, x + x_random, y + y_random)
print_state(env.get_heatmap())
env.step()
print_state(env.get_heatmap())
```
